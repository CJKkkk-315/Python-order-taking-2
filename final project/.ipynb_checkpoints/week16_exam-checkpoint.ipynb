## 3. 用户评论发表时间分析
（1）一天有24个小时，试统计每个时间点（小时）的评论总数，并按评论总数对时间点进行排序。试分析，用户一般在哪些时间点（小时）频繁或不经常发表评论？

（2）构建一个新的DataFrame对象（16 rows，命名为`top15_others`），包含排名前$15$的每小时评论总数、及其余小时（其余小时命名为`'Others'`）评论总和。并绘制饼图。


```python
time = [int(str(i).replace(':',' ').split(' ')[1]) for i in df['review_time'].tolist()]
d = Counter(time)
xy = [[i,j] for i,j in d.items()]
xy.sort(key=lambda x:x[1],reverse=True)
for i in xy:
    print(i[0],i[1])
xy15 = xy[:15]
xy16 = xy[15:]
other = sum([i[1] for i in xy16])
x = [str(i[0]) for i in xy15]
x.append('other')
y = [i[1] for i in xy15]
y.append(other)
d = {'cols':y}
top15_others = pd.DataFrame(d)
print(top15_others)
plt.pie(labels=x,x=y)
```
